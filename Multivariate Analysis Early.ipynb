New considerations when going from simple to multiple regression
<ol>
    <li> 
        Adding more independent variables to a multiple regression does not mean that the regression is better or that it will offer better predictions, in fact it can be worse. This is called OVERFITTING
    </li>
    <li>
        The addition of more independent variables creates more relationships among them. So not only do we need to consider the independent variables and the dependent variable, we also need to consider the relationships between the independent variables themselves. This is called MULTICOLINEARITY
    </li>
    <li>
        The ideal of for all independent variables to be correlated with the dependent variable and NOT with each other
    </li>
    <li>
        Because of overfitting and multicolinearlity, there is a lot of prep-work to before we start multi-regression.
        <ul>
            <li> Scatterplots </li>
            <li> Correlations </li>
            <li> Simple regressions between each of the independent variables and the dependent variables to see how they are related
        </ul>
    </li>
</ol>

Important to consider which independent variable make the cut.
Some independent or sets of independent variables are better at predicting the dependent variable than others. Some contribute nothing.

Multiple regression model.
\$ \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_p x_p + \epsilon $

\$ y = 27 + 9x_1 +12x_2 $

<ul>
    <li>$x_1$ = capital investment (€1000s)</li>
    <li>$x_2$ = marketing expenses (€1000s)</li>
    <li>y = predicted sales (€1000's)</li>
</ul>

Each coefficient (9, 12) is interpreted as the estimate in y corresponding to  a 1 unit change in variable when all other variables are held fixed.

€9000 is an estimate of the expected increase in sales y, corresponding to a €1000 increase in capital investment x1 when marketing expenditures x2 are kept fixed.

Conducting multiple regression requires the following steps:
<ol>
    <li> Generate a list of independent and dependent variables</li>
    <li> Collect data on the variables </li>
    <li> Check for outliers </li>
    <li> Check the relationships between each independent variable and the dependent variable using scatterplots and correlations</li>
    <li> Check the relationships between the independent variables using scatterplots and correlations </li>
    <li> Conduct simple linear regressions for each IV/DV pair</li>
    <li>Use the non-redundant independent variables in the analysis to find the best fitting model </li>
    <li> Use the best fitting model to make predictions about the dependent variable </li>
    </ol>

```python
corr = pearsonr(miles_travelled, time_taken)
print(corr)
```

First value is the correlation (very high) in this case. Second value is the p-value. That is probability that this is a random occurrence.

If the p-value is below 0.05 we say that it is statistically significant.

```python
model = LinearRegression()
miles_travelled = np.array(miles_travelled).reshape(-1, 1)
model.fit(miles_travelled, time_taken)
r_sq = model.score(miles_travelled, time_taken)
print(r_sq)
```

<ul>
<li>R-squared is the % of the variation in the dependent variable explained by the independent variable.</li>
<li>Standard error (SE) of the regression is the average distance of the points away from the regression line.</li>

```python
import statsmodels.api as sm
```

```python
miles_travelled = [89, 66, 78, 111, 44, 77, 80, 66, 109, 76]
miles_travelled = sm.add_constant(miles_travelled)
model = sm.OLS(time_taken, miles_travelled)
result = model.fit()
print(result.summary())
# Line below is for the standard error
print(result.scale**0.5)
```

```python
number_deliveries = sm.add_constant(number_deliveries)
model = sm.OLS(time_taken, number_deliveries)
result = model.fit()
print(result.summary())
# Line below is for the standard error
print(result.scale**0.5)
```

```python
fuel_prices = sm.add_constant(fuel_prices)
model = sm.OLS(time_taken, fuel_prices)
result = model.fit()
print(result.summary())
# Line below is for the standard error
print(result.scale**0.5)
```

```python
fuel_prices = [3.84, 3.19, 3.78, 3.84, 3.57, 3.57, 3.03, 3.51, 3.54, 3.24]
data = list(zip(miles_travelled, fuel_prices))
print(data)
```

```python
data = sm.add_constant(data)
model = sm.OLS(time_taken, data)
result = model.fit()
print(result.summary())
# Line below is for the standard error
print(result.scale**0.5)
```

The model says that if miles_travelled is held constant, then the travel time is expected to decrease by 0.2065 hours for each additonal dollar increase in fuel. This makes no sense. We should eliminate variables that are not correlated with the dependent variable as it messes up the coefficients in the regression.

```python
data = list(zip(number_deliveries, fuel_prices))
print(data)
```

```python
data = sm.add_constant(data)
model = sm.OLS(time_taken, data)
result = model.fit()
print(result.summary())
# Line below is for the standard error
print(result.scale**0.5)
```

Consider the model above. If deliveries are held constant, then travel_time us expected to decrease by 0.7332 of an hour for every additonal dollar increase in fuel_prices. This makes no sense. The p-value of 0.15 (>>0.05) indicates that this is going to be an issue.

```python
all_data = list(zip(miles_travelled, number_deliveries, fuel_prices))
print(all_data)
```

```python
all_data = sm.add_constant(all_data)
model = sm.OLS(time_taken, all_data)
result = model.fit()
print(result.summary())
# Line below is for the standard error
print(result.scale**0.5)
```

How do we select the best model? Consider several factors:
<ol>
    <li> $R^2$ adjusted </li>
    <li> The standard error </li>
    <li> The AIC and BIC as low as possible </li>
    <li> Consider the p-values for all coefficients in the model. If they are not all significant this indicates a problem </li>
    <li> Consider the model. Does it make sense. Often if in the point above coefficents are not significant the model can be can not make sense when you consider the problem </li>
    <li> If the Cond No. is very high this is a good indicator of problems </li>
    <li> All else being equal the simplest model is the best </li>
</ol>

<b> Total sum of squares </b>
<ul>
    <li>SST is the sum of the squared distances from the mean model (a horizontal line having slope 0 at the mean of the prices) and each observation.</li>
    <li> Once the data is locked in w.r.t the total number of observation (row) the total sum of squares is set </li>
    <li> SST is a finite sum that does not change regardless of how many features are part of the regression model: SST is not feature dependent it is row dependent </li>
    <li> Feature variables partition (allocate) the total sum of squares to the model and the remaining sum of squares are placed in error (redisuals)</li>
    <li> As features are added or removed the partitioning of the SSE and SSR will almost certainly change </li>
    <li> How the sum of squares is allocated to each feature within SSR will change as model changes </li>
    <li> If no features have significant explanatory power on y, the model will have zero features, and the best prediction will always be $\bar{y}$, the horizontal line with slope 0 </li>
    

SST is split into SSR (Regression model sum of squares) and SSE (residual error sum of squares). 

$R^2 = SSR/SST$ it describe the proportion of the sum of squares that we can put in the model. The goal is to:
<ul>
    <li> Maximise SSR as a proportion of SST</li>
    <li> Minimise SSE as a proportion of SST</li>
</ul>

<ul>
    <li> SSR is then split among the features in the model in terms of relative explanatory power </li>
    <li> Not only is SSR likely to change but the allocation of SSR to each features in the model changes as features are added and removed</li>
    <li> The SST is set regardless of the number of features in the model</li>
<ul>

<b> Overfitting and SSR </b>
<ul>
    <li> As more features are added to the model SSR will never decrease; SSR can only stay the same or increase </li>
    <li> As additional features are added, the model can begin to shape itself to the data it is based on (trined on), keeps optimising </li>
    <li> The model seeks to squeeze out as much SSE as possible whether the features add predictive power or not </li>
    <li> Complex regressions could suffer on new data because the nooks and crannies of the training data are absent </li>
    <li> There are techniques to avoid overfitting:</li>
        <ul>
            <li> Model building procedures (Forward selection, Backward elimination, Stepwise regression, Best subsets regression) along with  AIC, BIC, BC and Mallows C_p. </li>
            <li> Regularisation penalises complex models </li>
            <li> Cross validation </li>
            <li> PCA Principal Component Analysis </li>
        </ul>

<b> Multiple Regression -> Forward Selection </b>
There are three common techiques for building iterative regression models:
<ol>
    <li> Forward Selection </li>
    <li> Backward Elimination </li>
    <li> Stepwise Regression </li>
<ol>

There is fourth common technique is not iterative. It is called best subsets regression. Best subsets examine all possible combinations of feature variables; brute force method that can computationally expensive.
<ul>
    <li> Best subsets will examine all of the options, possibly 100's</li>
    <li> The analysts can specify the maximum number of features </li>
    <li> The analyst can also request the best two or three models for each number of feature variables </li>
</ul>

<b> Forward Selection </b>
<ul> Feature variables are examined one at a time </ul>
<li> The analyst sets a stopping rule (p-value) that acts as a hurdle the variable has to overcome to get into the model </li>
<li> The added variable must reduce the error (SSE) significantly </li>
<li> The feature variable that reduces the model error (SSE) the most is chosen as long as it passes the stopping </li>
<li> Once a variable is added to the model, it stays. It is never removed</li>
<li> The process repeats itself until all variables are in the model OR no variable passes the stopping rule </li>


<b> Flaws with forward selection </b>
<ul>
    <li> The ability of variables to reduce error can change as other variables enter the model </li>
    <li> As feature variables are added they begin to overlap and interact in how they explain variance </li>
    <li> Since a feature variable stays in the model once it is selected, forward selection is not flexible </li>
    <li> It is possible for a variable that entered early to fail the stopping rule later as other variables are added; but is is stuck in the model</li>
    <li> Example: $V1$ enters first. Then $V2$. Then $V3$. Let's say that $V2$ and $V2$ combined reduce error better than $V1$ alone. This could lead to $V1$ failing the stopping rule, but it is stuck in the model. </li>
    <li> Squared semipartial correlations can tease out these relationships </li>
    <li> It can miss suppressor and/or complimentary relationships:
        <ul>
            <li> Suppressor variables(also called enhancer variables) are strange. They correlate with parts of other variables but not the target. They eat up (suppress) some of the non-relevant error in those variables making the model stronger </li>
            <li> Complimentary variables are negatively correlated with each other but together they explain the target variable well; tradeoff.</li>
</ul>
